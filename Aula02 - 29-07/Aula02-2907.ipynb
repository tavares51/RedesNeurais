```python
import numpy as np
```

```python
X = [1.4, -0.7, 0.6, -0.5]
Y = [1, 0, 1, 1]
Θ = [0.0, 0.0]
```

```python
hTheta = []
theta_0 = Θ[0]
theta_1 = Θ[1]
for x in X:
    z = theta_0 + theta_1*x
    hTheta.append(1/(1+np.exp(-1 * z)))
```

```python
hTheta
```

```python
SUM_COST = 0
for i in range(len(Y)):
    COST = -Y[i] * np.log(hTheta[i]) - (1 -Y[i]) * np.log(1 - hTheta[i])
    SUM_COST += COST
```

```python
SUM_COST / 4
```
