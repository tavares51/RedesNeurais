```python
def perceptron(X, dT, alpha):
    # Inicializa os pesos (theta) com zeros, considerando o termo de bias
    theta = [0.0, -0.5, 0.5]

    n = 1  # Inicializa o número de iterações (épocas)

    while True:
        misclassified = False  # Verifica se ocorreu alguma classificação incorreta

        for x, d in zip(X, dT):
            # Adiciona o termo de bias (x0 = 1) no início do vetor de atributos
            x_with_bias = [1.0] + x

            # Calcula a ativação do neurônio (saída)
            y = activation(theta, x_with_bias)

            # Calcula o erro
            e = d - y

            if e != 0:  # Se houver erro na classificação
                misclassified = True

                # Ajuste de pesos
                for i in range(len(theta)):
                    theta[i] += alpha * e * x_with_bias[i]

        if not misclassified:  # Se não houver erro de classificação, encerra o loop
            break

        n += 1  # Incrementa o número de iterações

    return theta, n


def activation(theta, x):
    # Calcula a ativação do neurônio (saída)
    activation_sum = sum([w * xi for w, xi in zip(theta, x)])

    # Função de ativação (φ) - Heaviside step function
    return 1 if activation_sum >= 0 else -1
```
