## Implementação da método Perceptron

Antes de implementar o Perceptron, vamos separar os dados em treino (80%) e teste (20%).

```python
# define a porcentagem de treinamento
porcTrain = 0.8

# vamos usar uma semente para que sempre separe igualmente
randomSeed = 20

# variáveis que deverão guardar as classes de treino e teste
Y_train = None
Y_test = None

########################## COMPLETE O CÓDIGO AQUI  ########################
#
# (1) separe os dados em treino e teste usando a função StratifiedShuffleSplit da
#     biblioteca scikit learn
# (2) na inicializacao do modelo, configure o parâmetro random state com a semente informada
#     no código acima
# (3) use o método split da função StratifiedShuffleSplit para obter os índices de treino e teste.
#     Esse método retorna um generator. Para receber os valores, você pode usar um laço de
#     repetição ou usar a função next
#

from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1, train_size=porcTrain, random_state=0)

train_index, test_index = next(sss.split(X,Y))
X_train = X[train_index, :]
X_test = X[test_index, :]

Y_train = Y[train_index]
Y_test = Y[test_index]



##########################################################################

print('Qtd. dados de treinamento: %d (%1.2f%%)' %(X_train.shape[0], (X_train.shape[0]/X.shape[0])*100) )
print('Qtd. de dados de teste: %d (%1.2f%%)' %(X_test.shape[0], (X_test.shape[0]/X.shape[0])*100) )
```

Vamos treinar um modelo de classificação usando o método **perceptron**. Para isso, iremos empregar a biblioteca **scikit-learn**.

```python
# variável que deverá recer o modelo
model = None

########################## COMPLETE O CÓDIGO AQUI  ########################
#
# (1) importe o método Percetron da biblioteca Scikit-learn
# (2) inicialize o parâmetro referente à taxa de aprendizado com valor igual a 0.5 (eta0)
# (3) inicialize o método com o parâmetro referente ao máximo de iterações igual a 500 (max_iter)
# (4) inicialize o parâmetro que contém o critério de parada com valor igual a 0.0002 (tol)
# (5) inicialize o método com um valor para o parâmetro random state
# (6) use o método fit para treinar o modelo

from sklearn.linear_model import Perceptron

model = Perceptron(eta0=0.5, max_iter=500, tol = 0.0002, random_state= 0)
model.fit(X = X_train, y = Y_train)


##########################################################################

print('Informações do modelo: ')
print(model)


print('\nBIAS: ')
print(model.intercept_)

print('\nPesos encontrados: ')
print(model.coef_)

print('\nEpocas: ')
print(model.n_iter_)

print('\Qtd de Atulizacao de peso: ')
print(model.t_)
```

Vamos gerar a curva de aprendizado para ver se houve problema de overfitting ou underfitting.

```python
# valores que você precisa retornar
train_sizes = None,
train_scores = None,
test_scores = None

########################## COMPLETE O CÓDIGO AQUI  ########################
#
# (1) copie o código do perceptron usado anteriormente
# (2) inicialize a função StratifiedShuffleSplit e faça 10 divisoes dos seus dados de treino, com 20% para teste
# (3) use a função learning_curve para criar a curva de aprendizado.
#      Para o parametro trainsize, use o método linspace
# (4) inicialize o parâmetro que contém o critério de parada com valor igual a 0.0002 (tol)
# (5) inicialize o método com um valor para o parâmetro random state
# (6) use o método fit para treinar o modelo

from sklearn.model_selection import learning_curve
from sklearn.metrics import make_scorer, f1_score

classifier = Perceptron(eta0=0.1, max_iter=300, tol = 0.01, random_state= 0)
medida = make_scorer(f1_score, average = 'macro')
modelCv = StratifiedShuffleSplit(n_splits=10, train_size=0.8, random_state=0)
train_sizes,train_scores,test_scores =  learning_curve(estimator = classifier,
                            X = X_train,
                            y = Y_train,
                            train_sizes = np.linspace(0.1, 1, 10),
                            scoring = medida,
                            cv = modelCv,
                            random_state=0
                            )


##########################################################################

# calcula a media e desvio padrão
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)

test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

plt.figure(figsize=(8,6))
plt.plot(train_sizes, train_mean, '--', color="black",  label="Training score")
plt.plot(train_sizes, test_mean, color="black", label="Cross-validation score")

plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="cyan")
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="cyan")

plt.title("Curva de aprendizado")
plt.xlabel("Tamanho do conjunto de treinamento"),
plt.ylabel("Acurácia"),
plt.legend()
plt.show()
```

Agora, vamos testar os pesos aprendidos pelo Perceptron na classificação dos dados de teste.

```python
# inicializa o vetor com as predições
Y_pred = np.zeros( len(Y_test) )

########################## COMPLETE O CÓDIGO AQUI  ########################
#
# (1) use o método predict para fazer as predições no conjunto de teste
#

Y_pred = model.predict(X_test)


##########################################################################


print("Cinco primeiras classes preditas")
print(Y_pred[0:5])
```

Vamos criar uma funcao para plotar a superfície de decisão criada pelo Perceptron.

```python
#função para plotar o limite de decisão
def plota_limite_decisao(model, X, Y, ax):

    x = X[:, 0]
    y = X[:, 1]

    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),np.arange(y_min, y_max, 0.01))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = np.array(Z)
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

    ax.set_xlim(x.min()-0.1, x.max()+0.1)
    ax.set_ylim(y.min()-0.1, y.max()+0.1)
    ax.set_xticks(())
    ax.set_yticks(())

    #plota os dados da classe 0
    plt.scatter( X[Y==0,0], X[Y==0,1], label='Classe 0', color='blue', s=50, edgecolors='k')

    #plota os dados da classe 1
    plt.scatter( X[Y==1,0], X[Y==1,1], label='Classe 1', color='red', s=50, edgecolors='k')

    #insere a legenda
    plt.legend()


# define o tamanho da figura
fig, ax = plt.subplots(figsize=(8, 8))

# plota o limite de decisão
plota_limite_decisao(model, X_train, Y_train, ax)
plt.show()
```

## Aprendizado Online

O Perceptron é um método de aprendizado online. Portanto, é possível atualizar o modelo de treinamento com um ou mais dados.

Execute o código abaixo para carregar alguns dados novos.

```python
X_novo = np.array([
    [4.91,3.05],
    [5.50,2.30],
    [6.05,3.05],
    [5.05,3.32]])

Y_novo = np.array([1,1,0,0])

print('Novos dados carregados com sucesso!')
```

Use os dados acima para atualizar o algoritmo. O algoritmo permite que sejam forcecidos todos os dados de uma vez, mas em vez disse passe um único dado por vez usando a estratégia de **treinar apenas quando o modelo errar a predição (train-on-error)**. Essa estratégia é muito usada em classificação de spam. Depois de cada atualização, imprima o parâmetro que mostra a quantidade de atualizações de pesos do Perceptron.

```python
print('Qtd. de atualizações dos pesos até o momento: \n', model.t_)

########################## COMPLETE O CÓDIGO AQUI  ########################
#
# (1) Use o método partial_fit para fazer o aprendizado online do perceptron

for i in range(len(Y_novo)):

    y_pred = model.predict([X_novo[i,:]])
    print('Dado novo: %d - predita: %d' % (i, y_pred))
    if y_pred != Y_novo[i]:
        print("\t Errou\n")
        model.partial_fit([X_novo[i,:]], [Y_novo[i]])



##########################################################################
```
