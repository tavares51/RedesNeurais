```python
def calculaJ(X, Y, formula):
    if len(formula) == 0:
        raise "Formula não inserida"

    if len(X) == 0:
        raise "X não inserido"

    if len(Y) == 0:
        raise "Y não inserido"

    hTeta = []
    teta_0 = formula[0]
    dist = []

    for x in X:
        x_aux = []
        for teta in formula[1:]:
            x_aux.append(teta*x)

        hTeta.append(teta_0 + sum(x_aux))

    for y, h in zip(Y, hTeta):
        dist.append(pow(y - h, 2))

    j = sum(dist) / (len(Y) * 2)
    return j
```

```python
def calculaJ(X, Y, formula):
    if len(formula) == 0:
        raise "Formula não inserida"

    if len(X) == 0:
        raise "X não inserido"

    if len(Y) == 0:
        raise "Y não inserido"

    teta_0 = formula[0]
    dist = []

    hTeta = np.dot(formula, X)

    for y, h in zip(Y, hTeta):
        dist.append(pow(y - h, 2))

    j = sum(dist) / (len(Y) * 2)
    return j
```
