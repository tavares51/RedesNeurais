# <center>Regressão linear</center>

Neste exemplo, iremos criar um problema de regressão aleatório.

```python
import sklearn as skl
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import datasets

import numpy as np

n_samples=200
X, Y, coef = datasets.make_regression(n_samples=n_samples, n_features=1,
                                      n_informative=1, noise=30,
                                      coef=True, random_state=10)

# imprime as 5 primeiras linhas da matriz X
print('Primeiros valores de X:', X[0:5])

# imprime os 5 primeiros valores de Y
print('Primeiros valores de Y:', Y[0:5])
```

Agora, vamos criar uma função para plotar os dados usando um gráfico de dispersão.

```python
# importando a biblioteca matplotlib
import matplotlib.pyplot as plt

#criando a função para plotar os dados
def visualizarDados(X,Y):
    """
    Função usada para plotar os dados

    PLOTARDADOS(X,Y) plota pontos de dados e seta os rotulos x e y da
    figura com "Populacao" e "Orcamento", respectivamente.
    """



#chamando a função que plota os dados
visualizarDados(X,Y)

# mostra o gráfico usando a função show() da matplotlib
plt.show()
```

Vamos dividir os dados em dois subconjuntos: treino (80%) e teste (20%).

```python
X_train = None
X_test = None
Y_train = None
Y_test = None

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) use o método train_test_split do scikit learn e divida os dados em 80/20
# (2) inicialize o método com um valor para o parâmetro random state
#

X_train, X_test, Y_train, Y_test = skl.model_selection.train_test_split(
      X, Y, train_size = 0.8, random_state = 10)

##########################################################################

print('Qtd. dados de treinamento: %d (%1.2f%%)' %(X_train.shape[0], (X_train.shape[0]/X.shape[0])*100) )
print('Qtd. de dados de teste: %d (%1.2f%%)' %(X_test.shape[0], (X_test.shape[0]/X.shape[0])*100) )
```

Vamos inicializar e treinar o modelo de regressão Linear.

```python
model = None

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) importe o método de Regressão Linear da biblioteca Scikit-learn
# (3) inicialize o método com um valor para o parâmetro random state
# (4) use o método fit para treinar o modelo

# inicializa o modelo
#model = LinearRegression()
model = skl.linear_model.Ridge(alpha=0.0001,
                               max_iter = 1,
                               random_state=10,
                               tol=0)

# treina o algoritmo
model.fit(X_train, Y_train)

##########################################################################

print('Informações do modelo: ')
print(model)

print('\nBias: ', model.intercept_)

print('\nPesos encontrados: ')
theta = model.coef_ # returns a matrix of weights (coefficients)
print(theta)
```

Vamos dar uma olhada no regressor linear aprendido.

```python
# plota os dados de treinamento
visualizarDados(X_train,Y_train)

# plota os dados
plt.plot(X_train, np.dot(X_train,theta), label = 'Regressor Linear', color='red', linestyle='-', linewidth=1.5)

# Plota a legenda
plt.legend()

plt.show()
```

Agora, teste o modelo.

```python
Y_pred = None

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) use o método predict para fazer a predição

# faz a predicao nos dados de teste
Y_pred = model.predict(X_test)

##########################################################################

print("Mostra a predição para as 10 primeiras amostras")
print(Y_pred[0:10])
```

Calcula o desempenho do modelo.

```python
r2 = 0.0
mae = 0.0
mse = 0.0

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) use o método r2_score da biblioteca scikit learn para calcular a métrica R2
# (2) use o método mean_absolute_error da biblioteca scikit learn para calcular o erro médio absoluto
# (3) use o método mean_squared_error da biblioteca scikit learn para calcular o erro quadrático médio
#

r2 = skl.metrics.r2_score(Y_test, Y_pred)
mae = skl.metrics.mean_absolute_error(Y_test, Y_pred)
mse = skl.metrics.mean_squared_error(Y_test, Y_pred)

##########################################################################

print('R2: %1.2f' %r2)
print('MAE: %1.2f' %mae)
print('MSE: %1.2f' %mse)
```

## Regressão linear de um problema mais complexo

```python
X = np.sort(5 * np.random.rand(40, 1), axis=0)
Y = np.sin(X).ravel()

# imprime as 5 primeiras linhas da matriz X
print('Primeiros valores de X:', X[0:5])

# imprime os 5 primeiros valores de Y
print('Primeiros valores de Y:', Y[0:5])
```

Viasualize os dados

```python
#chamando a função que plota os dados
visualizarDados(X,Y)

# mostra o gráfico usando a função show() da matplotlib
plt.show()
```

Vamos dividir os dados em dois subconjuntos: treino (80%) e teste (20%).

```python
X_train = None
X_test = None
Y_train = None
Y_test = None

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) use o método train_test_split do scikit learn e divida os dados em 80/20
# (2) inicialize o método com um valor para o parâmetro random state
#

X_train, X_test, Y_train, Y_test = skl.model_selection.train_test_split(
      X, Y, train_size = 0.8, random_state = 10)

##########################################################################

print('Qtd. dados de treinamento: %d (%1.2f%%)' %(X_train.shape[0], (X_train.shape[0]/X.shape[0])*100) )
print('Qtd. de dados de teste: %d (%1.2f%%)' %(X_test.shape[0], (X_test.shape[0]/X.shape[0])*100) )
```

Crie atributos polinomias. Teste diferentes graus de polinômio.

```python
########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) importe o método PolynomialFeatures da biblioteca Scikit-learn
# (3) gere features polinomiais com diferentes graus  de polinomio
# (4) use o método fit para treinar o modelo

from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=5, include_bias=False)

X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

print(X_train_poly)
```

Vamos inicializar e treinar o modelo de regressão Linear.

```python
model = None

########################## COMPLETE O CÓDIGO AQUI  ########################
# (1) importe o método de Regressão Linear da biblioteca Scikit-learn
# (3) inicialize o método com um valor para o parâmetro random state
# (4) use o método fit para treinar o modelo

# inicializa o modelo
model = LinearRegression()

# treina o algoritmo
model.fit(X_train_poly, Y_train)

##########################################################################

print('Informações do modelo: ')
print(model)

print('\nPesos encontrados: ')
theta = model.coef_ # returns a matrix of weights (coefficients)
print(theta)
```

Vamos dar uma olhada no regressor linear aprendido.

```python
# usa para gerar a linha aprendida pelo modelo
maxTrain = max(X_train)
minTrain = min(X_train)

ux = np.linspace(minTrain, maxTrain, 50)
ux_poly = poly.transform(ux)

# plota os dados de treinamento
visualizarDados(X_train,Y_train)

# plota os dados
plt.plot(ux, np.dot(ux_poly,theta), label = 'Regressor Linear', color='red', linestyle='-', linewidth=1.5)

# Plota a legenda
plt.legend()

plt.show()
```
